```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

%matplotlib inline
plt.rc('font',family='Times New Roman') 
plt.style.use("ggplot")
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

font_T = fm.FontProperties(family='Times New Roman', size=24, stretch=0, weight='bold')
```

```python
letter_count_position = pd.read_csv('letter_count_position.csv', index_col=0)
letter_count_position
```

```python
sns.set_theme(style="white")  
# 
fig, ax = plt.subplots(figsize=(16, 9))

x = letter_count_position.index
y1 = letter_count_position['Position1'].values
y2 = letter_count_position['Position2'].values
y3 = letter_count_position['Position3'].values
y4 = letter_count_position['Position4'].values
y5 = letter_count_position['Position5'].values

plt.bar(x, y1, 
        label='Position1', alpha=0.75)
plt.bar(x, y2, bottom=y1,
        label='Position2', alpha=0.75)
plt.bar(x, y3, bottom=y1+y2,
        label='Position3', alpha=0.75)
plt.bar(x, y4, bottom=y1+y2+y3,
        label='Position4', alpha=0.75)
plt.bar(x, y5, bottom=y1+y2+y3+y4,
        label='Position5', alpha=0.75)

# 
plt.title('Letter Frequency by Position', fontproperties = font_T)
plt.xlabel('Letter', fontproperties = font_T)
plt.ylabel('Number', fontproperties = font_T)

# 
plt.xticks(fontproperties = {'family':'Times New Roman', 'size':20})
plt.yticks(fontproperties = {'family':'Times New Roman', 'size':15})
plt.tick_params(axis='both', bottom='on', left='on')
plt.legend(prop = {'family':'Times New Roman', 'size':18})

plt.tight_layout()
plt.savefig('letter_count_position.png', dpi=300, bbox_inches='tight')
```
